```python

#link for song list page.

for l in List_AB:
    linksub = l
    link = head_link + linksub

    #Open the webpage.
    driver.get(link)
    page_content = driver.page_source
    #Pulling entire page data.
    pulled_content = BeautifulSoup(page_content)
    #Pulling only required content.
    album_link = pulled_content.findAll('div', attrs = {'class':'one-third column'})
    album_link = str(album_link)
    album_links = re.findall(r'[/][l][y][r][i][c][s].[\w]+.+.[h][t][m][l]', album_link)
    #album_links


    #Iterating each album and pulling links for each song from that specific album.
    song_links = []
    for al in album_links:
        al_full = head_link + al
        #print(al_full)
        #Open the webpage.
        driver.get(al_full)
        page_content = driver.page_source
        #Pulling entire page data.
        pulled_content = BeautifulSoup(page_content)
        #Pulling only required content.
        song_link = pulled_content.findAll('ul', attrs = {'class':'square'})
        song_link = str(song_link)
        ind_song_links = re.findall(r'[/][l][y][r][i][c][s].[\w]+.+.[h][t][m][l]', song_link)
        #print(ind_song_links)
        song_links.append(ind_song_links)
    

```

```python

#Iterating each song and pulling lyrics into a list.
lyrics_list = []
for asl in song_links:
    for sl in asl:
        sl_full = head_link + sl
        print(sl_full)
        #Open the webpage.
        driver.get(sl_full)
        page_content = driver.page_source
        #Pulling entire page data.
        pulled_content = BeautifulSoup(page_content)
        #Pulling only required content.
        lyrics = pulled_content.findAll('font', attrs = {'face':'verdana'})
        lyrics_list.append(lyrics)

#Writing the lyrics in a csv file!
file = open('songs_AB.csv','w+', encoding = 'utf-8', newline = '\n')
with file:
    objects = csv.writer(file)
    objects.writerows(lyrics_list)
file.close()

```
